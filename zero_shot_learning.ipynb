download and install bert model from huggingface pipeline and create an executable object of it.

```python
import gradio as gr
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

Define a function to execute the bart model with the given input and return the output.

```python
def aiMagic(sentence, intents):
  return classifier(sentence, intents)
```

define a gradio function and get the inputs from the user, call the aiMagic to calculate the output and after then return the output.

```python
demo = gr.Interface(
    fn=aiMagic,
    inputs=["text", "text"],
    outputs=["json"],
)
```

Lanch the gradio app

```python
demo.launch()
```
